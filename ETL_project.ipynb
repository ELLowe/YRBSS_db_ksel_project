```python
import pandas as pd
from sqlalchemy import create_engine
alcohol_file = "Alcohol_and_Other_Drug_Use.csv"
alcohol_df = pd.read_csv(alcohol_file)
```

```python
alcohol_df.head()
```

```python
alcohol_df.columns
```

```python
cleaned_alcohol_df = alcohol_df[["YEAR","LocationDesc","Subtopic","Description","Greater_Risk_Question","Greater_Risk_Data_Value","Lesser_Risk_Question","Lesser_Risk_Data_Value","Sample_Size","Sex","Race","Grade","LocationId","GeoLocation","StratificationType"]]
cleaned_alcohol_df = cleaned_alcohol_df.dropna()
cleaned_alcohol_df= cleaned_alcohol_df.reset_index(drop=True)
cleaned_alcohol_df=cleaned_alcohol_df.rename(columns={"YEAR":'surv_year'})
cleaned_alcohol_df.head()
```

```python
tobacco_file= "Tobacco_Use.csv"
tobacco_df=pd.read_csv(tobacco_file)
```

```python
tobacco_df.head()
```

```python
tobacco_df.columns
```

```python
cleaned_tobacco_df.dtypes
```

```python
cleaned_tobacco_df = tobacco_df[["YEAR","LocationDesc","Subtopic","Description","Greater_Risk_Question","Greater_Risk_Data_Value","Lesser_Risk_Question","Lesser_Risk_Data_Value","Sample_Size","Sex","Race","Grade","LocationId","GeoLocation","StratificationType"]]
cleaned_tobacco_df = cleaned_tobacco_df.dropna()
cleaned_tobacco_df= cleaned_tobacco_df.reset_index(drop=True)
cleaned_tobacco_df=cleaned_tobacco_df.rename(columns={"YEAR":'surv_year'})
cleaned_tobacco_df.head()
```

```python
Sexual_Behaviors_file= "Sexual_Behaviors.csv"
Sexual_Behaviors_df=pd.read_csv(Sexual_Behaviors_file)
```

```python
Sexual_Behaviors_df.head()
```

```python
cleaned_Sexual_Behaviors_df.columns
```

```python
cleaned_Sexual_Behaviors_df = Sexual_Behaviors_df[["YEAR","LocationDesc","Subtopic","Description","Greater_Risk_Question","Greater_Risk_Data_Value","Lesser_Risk_Question","Lesser_Risk_Data_Value","Sample_Size","Sex","Race","Grade","LocationId","GeoLocation","StratificationType"]]
cleaned_Sexual_Behaviors_df = cleaned_Sexual_Behaviors_df.dropna()
cleaned_Sexual_Behaviors_df= cleaned_Sexual_Behaviors_df.reset_index(drop=True)
cleaned_Sexual_Behaviors_df=cleaned_Sexual_Behaviors_df.rename(columns={"YEAR":'surv_year'})
cleaned_Sexual_Behaviors_df.head()
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/YRBSS"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
cleaned_alcohol_df.to_sql(name='alcohol', con=engine, if_exists='append', index=False)
```

```python
cleaned_tobacco_df.to_sql(name='tobacco', con=engine, if_exists='append', index=False)
```

```python
cleaned_Sexual_Behaviors_df.to_sql(name='sexual_behaviors', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from alcohol', con=engine).head()
```
